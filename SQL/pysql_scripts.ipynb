```python
from sqlalchemy import create_engine
import pandas as pd
```

```python


# DEFINE THE DATABASE CREDENTIALS
user = 'root'
password = 'root'
host = '127.0.0.1'
port = 3306
database = 'bigdataclass2025'

# PYTHON FUNCTION TO CONNECT TO THE MYSQL DATABASE AND
# RETURN THE SQLACHEMY ENGINE OBJECT
def get_connection():
    return create_engine(
        url="mysql+pymysql://{0}:{1}@{2}:{3}/{4}".format(
            user, password, host, port, database
        )
    )
    
try:
    # GET THE CONNECTION OBJECT (ENGINE) FOR THE DATABASE
    engine = get_connection()
    print(f"Connection to the {host} for user {user} created successfully.")
except Exception as ex:
    print("Connection could not be made due to the following error: \n", ex)
```

```python
df_sales = pd.read_sql('select * from sales', engine)
df_sales
```

```python
df_products = pd.read_sql('select * from products', engine)
df_products
```

```python
df_sales_products = pd.merge(df_sales, df_products, on='product_id')
df_sales_products
```

```python
df_sales_products1 = df_sales.merge(df_products, on='product_id')
df_sales_products1
```

```python
df_sales_products.to_sql('sales_products', engine, if_exists='replace', index=False)
```
